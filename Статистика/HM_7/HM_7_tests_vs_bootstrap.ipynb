Вроде наши данные не распределены нормально, давайте посмотри на график QQ-plot и тест Шапиро-Уилка, чтобы точно в это убедиться

```python
sm.qqplot(df_slivki['likes'], line = 's')
plt.title('Лайки у Сливки шоу')

sm.qqplot(df_syend['likes'], line = 's')
plt.title('Лайки у Сыендука')
```

График QQ-plot нам показал, что данные у нас не нормальные. Посмотрим на тесст Шапиро-Уилка

Да, данные у нас 100% не нормально распределены, а значит будем использовать тест Манна-Уитни (непараметрический аналог тесту Стбюдента)

```python
print(sts.mannwhitneyu(df_slivki['likes'], df_syend['likes']))
```

У нас p-value намного меньше 0.05, а значит, между выборками есть стат. значимая разница, а именно, у нас медианные значения лайков не совпадают и распределения являются разными сами по себе

## 1.4 -- Частота категорий видео у Happy Wolf и Сыендук

Под категорией видео я подразумеваю насколько хорошо видео зашло пользователем (допустим, это будет метрика отношение лайков к дизлайкам)

Здесь мы попробуем применить тест Кси-квадрат, так как у нас будут 3 категории видео:

* крутые
* средние
* не очень

```python
d = {0: 1}
df_wolf['dislikes'] = df_wolf['dislikes'].replace(d)
df_syend['dislikes'] = df_syend['dislikes'].replace(d)
df_slivki['dislikes'] = df_slivki['dislikes'].replace(d)

df_wolf['likes_to_dislikes'] = df_wolf['likes'] / df_wolf['dislikes']
df_syend['likes_to_dislikes'] = df_syend['likes'] / df_syend['dislikes']
df_slivki['likes_to_dislikes'] = df_slivki['likes'] / df_slivki['dislikes']

def assign_category(row, q1, q3):
    if row['likes_to_dislikes'] > q3:
        row['category'] = 'Крутое видео'
    elif row['likes_to_dislikes'] < q3 and row['likes_to_dislikes'] > q1:
        row['category'] = 'Среднее видео'
    else:
        row['category'] = 'Плохое видео'
    return row['category']


q1_wolf = np.percentile(df_wolf['likes_to_dislikes'], q=25)
q3_wolf = np.percentile(df_wolf['likes_to_dislikes'], q=75)

print(q1_wolf)
print(q3_wolf)


df_wolf['category'] = df_wolf.apply(assign_category, q1 = q1_wolf, q3 = q3_wolf, axis = 1)


q1_syend = np.percentile(df_syend['likes_to_dislikes'], q=25)
q3_syend = np.percentile(df_syend['likes_to_dislikes'], q=75)

df_syend['category'] = df_syend.apply(assign_category, q1 = q1_syend, q3 = q3_syend, axis = 1)

display(df_wolf['category'].value_counts().reset_index())
display(df_syend['category'].value_counts().reset_index())


wolf_slivki_table = [[36, 48],  # Среднее видео
                     [19, 25],  # Плохое видео
                     [19, 25]]  # Крутое видео


result = sts.chi2_contingency(wolf_slivki_table).pvalue
print(result)
```

Итак, мы получили p-value > 0.99, а это значит, что распределние категорий видео у нас одинаковое

**А в чем отличие теста Фишера и Кси-квадрат?**




Тест Кси-квадрат для больших данных (больше 30) и для >= 2 категорий, а тест Фишера только для маленьких показателей и 2 категорий, но **тест Фишера точнее**

## 2.3 -- Частота попадания видео в тренды у каналов Сливки Шоу и Сыендук 

```python
slivki_syend_table = np.array([[184, 126],
                                [56, 42]])

# Количество бутстрап-итераций
n_iterations = 1000
bootstrap_fisher_results = []

for _ in range(n_iterations):
    # Генерируем бутстрап-выборку для каждой группы
    
    boot_slivki_trend = np.random.binomial(n=slivki_syend_table[0, 0] + slivki_syend_table[0, 1],
                                           p=slivki_syend_table[0, 0] / (slivki_syend_table[0, 0] + slivki_syend_table[0, 1]))
    boot_slivki_normal = (slivki_syend_table[0, 0] + slivki_syend_table[0, 1]) - boot_slivki_trend

    boot_syend_trend = np.random.binomial(n=slivki_syend_table[1, 0] + slivki_syend_table[1, 1],
                                          p=slivki_syend_table[1, 0] / (slivki_syend_table[1, 0] + slivki_syend_table[1, 1]))
    boot_syend_normal = (slivki_syend_table[1, 0] + slivki_syend_table[1, 1]) - boot_syend_trend

    # Создаем новую бутстрап-таблицу
    boot_table = np.array([[boot_slivki_trend, boot_slivki_normal],
                            [boot_syend_trend, boot_syend_normal]])

    # Применяем тест Фишера
    fisher_stat, p_value = sts.fisher_exact(boot_table)
    bootstrap_fisher_results.append(fisher_stat)

# Вычисляем доверительный интервал для коэффициента отношения шансов
left_ci = np.percentile(bootstrap_fisher_results, 2.5)
right_ci = np.percentile(bootstrap_fisher_results, 97.5)

print(f'Оригинальный коэффициент отношения шансов (odds ratio): {sts.fisher_exact(slivki_syend_table)[0]}')
print(f'95% доверительный интервал для odds ratio: [{left_ci}, {right_ci}]')
```

Если у нас 1 принадлежит доверительному интервалу, то все хорошо и тест Фишера нам показал правду

## 2.4 -- Частота категорий видео у Happy Wolf и Сыендук

```python
import numpy as np
import scipy.stats as sts

# Оригинальная таблица
wolf_slivki_table = np.array([[36, 48],  # Среднее видео
                              [19, 25],  # Плохое видео
                              [19, 25]])  # Крутое видео

original_chi2 = sts.chi2_contingency(wolf_slivki_table)[0]
print(f'Оригинальная статистика: {original_chi2}')

bootstrap_res = []
n_bootstrap = 1000

for i in range(n_bootstrap):
    # Бутстрап: случайный выбор строк из таблицы (с возвращением)
    resampled_table = np.array([[np.random.choice(wolf_slivki_table[row], size=2, replace=True) for row in range(3)]])
    
    chi2_stat = sts.chi2_contingency(resampled_table)[0]
    bootstrap_res.append(chi2_stat)

# 95% доверительный интервал
left_line = np.percentile(bootstrap_res, 2.5)
right_line = np.percentile(bootstrap_res, 97.5)

print(f'95% доверительный интервал для статистики хи-квадрат: [{left_line}, {right_line}]')
```
